```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
from mytools import Video, get_cuts
```

# Video 0

```python
video_filename = 'videos/0.mp4'
timeline_filename = 'timelines/0.kdenlive'
```

```python
video = Video(video_filename)

print(f'video.length = {video.length}\nvideo.width = {video.width}\nvideo.height = {video.height}\n')

frame_index = 221
frame = video.get_frame(frame_index)

print(f'frame_index = {frame_index}')
print(f'frame.shape = {frame.shape}\nframe.dtype = {frame.dtype}')

plt.imshow(frame)
```

```python
cuts = get_cuts(timeline_filename)
```

```python
diff = [abs(matrix[1] - matrix[0]).sum() for matrix in tqdm(video.generate_matrices(2), total=video.length-2)]
```

```python
# Normalize values, deviding them by max possible
diff = np.array(diff) / (3*255*video.width*video.height)
diff.shape
```

```python
plt.figure(figsize=(18, 6))

plt.plot(diff)
plt.title(f'{video_filename}\nRGB gradient')
plt.xticks(cuts, rotation=90)
plt.grid()
plt.show()
```
