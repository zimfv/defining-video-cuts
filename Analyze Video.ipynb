```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
from mytools import Video
```

```python
filename = 'videos/0.mp4'
```

```python
video = Video(filename)

print(f'video.length = {video.length}\nvideo.width = {video.width}\nvideo.height = {video.height}\n')

frame_index = 221
frame = video.get_frame(frame_index)

print(f'frame_index = {frame_index}')
print(f'frame.shape = {frame.shape}\nframe.dtype = {frame.dtype}')

plt.imshow(frame)
```

```python
diff = [abs(matrix[1] - matrix[0]).sum() for matrix in tqdm(video.generate_matrices(2), total=video.length-2)]
```

```python
diff = np.array(diff) / (255*video.width*video.height)
diff.shape
```

```python
plt.figure(figsize=(12, 6))

plt.plot(diff)
plt.title('RGB gradient')

plt.show()
```
