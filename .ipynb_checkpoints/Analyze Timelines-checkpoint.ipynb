```python
import numpy as np
import matplotlib.pyplot as plt
from mytools import Video, get_cuts
```

# Looking for cuts

```python
timeline_file = 'timelines/0.kdenlive'
```

```python
cuts = get_cuts(timeline_file)
print(f'There are {len(cuts)} in file {timeline_file}:\n\n {str(cuts)[1:-1]}')
```

```python
video_filename = 'videos/0.mp4'
video = Video(video_filename)
```

```python
midles = cuts[:16]

fig, axs = plt.subplots(len(midles), 4)
fig.set_figheight(2*len(midles))
fig.set_figwidth(4*3)

for i in range(len(midles)):
    midle = midles[i]
    for k in range(4):
        frame_id = midle - 2 + k
        frame = video.get_frame(frame_id)
        axs[i, k].imshow(frame)
        axs[i, k].set_title(f'Frame {frame_id}')
fig.savefig('cuts.png')
```
